```python
from transformers import BertTokenizer
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
sample_text = "I like Natural Language Processing and I work with Python."
```

```python
tokens = tokenizer.tokenize(sample_text)
```

```python
tokens_ids = tokenizer.convert_tokens_to_ids(tokens)
```

```python
tokens_ids
```

```python
tokenizer.sep_token, tokenizer.sep_token_id
```

```python
tokenizer.cls_token, tokenizer.cls_token_id
```

```python
encoding = tokenizer.encode_plus(
    sample_text,
    max_length=30,
    add_special_tokens=True,
    return_attention_mask=True,
    return_tensors='pt'
)
```

```python
encoding
```

```python
from transformers import pipeline
```

```python
sa_pipe = pipeline("text-classification")
sa_pipe(sample_text)
```

```python
sa_pipe([sample_text, "I did not like the resturant at all", "I did not like the food but the venue was pretty nice"])
```

```python
zsc = pipeline("zero-shot-classification")
zsc(
    "This is a webinar that deals with machine learning and natural language processing",
    candidate_labels = ["politics", "business", "education"]
)
```

```python
zsc(
    "Tesla has released the Q2 results yesterday and the stock surged by 2%",
    candidate_labels = ["politics", "business", "education"]
)
```

```python
zsc(
    "AWS SageMaker is used to train and deploy machine learning models",
    candidate_labels = ["politics", "business", "education", "technology"]
)
```
